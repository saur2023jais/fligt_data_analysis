## Find out the Monthly Total chargeable weight  for each Sector and make a bar plot w.r.t. each sector


```python

# Extract the month from the FlightDate column and store it in a new column 'Month'
df['Month'] = pd.to_datetime(df['FlightDate']).dt.month

# Group the data by Sector and Month, and calculate the sum of ChargeableWeight for each group
grouped = df.groupby(['Sector', 'Month'])['ChargeableWeightEXE'].sum()

# Convert the grouped data to a DataFrame
df_grouped = grouped.reset_index()

# Loop through each unique Sector and plot a bar chart for monthly total chargeable weight
for sector in df_grouped['Sector'].unique():
    sector_data = df_grouped[df_grouped['Sector'] == sector]
    sector_data.plot(kind='bar', x='Month', y='ChargeableWeightEXE', color='blue')
    plt.xlabel('Month')
    plt.ylabel('Monthly Total Chargeable Weight')
    plt.title(f'Monthly Total Chargeable Weight by Month for Sector {sector}')
    plt.show()
```

```python

# Extract the year from the FlightDate column and store it in a new column 'Year'
df['Year'] = pd.to_datetime(df['FlightDate']).dt.year

# Group the data by Sector and Year, and calculate the sum of ChargeableWeight for each group
grouped = df.groupby(['Sector', 'Year'])['ChargeableWeightEXE'].sum()

# Convert the grouped data to a DataFrame
df_grouped = grouped.reset_index()

# Loop through each unique Sector and plot a bar chart for yearly total chargeable weight
for sector in df_grouped['Sector'].unique():
    sector_data = df_grouped[df_grouped['Sector'] == sector]
    sector_data.plot(kind='bar', x='Year', y='ChargeableWeightEXE', color='green')
    plt.xlabel('Year')
    plt.ylabel('Yearly Total Chargeable Weight')
    plt.title(f'Yearly Total Chargeable Weight by Year for Sector {sector}')
    plt.show()
```

```python
from scipy import stats

# Group the data by ProductName
grouped = df.groupby('ProductName')

# Iterate through the groups and remove the outliers using z-score
for group_name, group in grouped:
    z_scores = np.abs(stats.zscore(group['ChargeableWeightEXE']))
    threshold = 3
    group_without_outliers = group[z_scores < threshold]
    df = df.drop(group.index.difference(group_without_outliers.index))

# Print the updated dataframe
print(df)

```

```python
# Group the data by Sector
grouped = df.groupby('Sector')

# Iterate through the groups and remove the outliers using z-score
for group_name, group in grouped:
    z_scores = np.abs(stats.zscore(group['ChargeableWeightEXE']))
    threshold = 3
    group_without_outliers = group[z_scores < threshold]
    df = df.drop(group.index.difference(group_without_outliers.index))

# Print the updated dataframe
print(df)
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing

# Step 1: Create a dataframe with required columns
df_od = df[['FlightDate', 'Origin', 'Destination', 'ChargeableWeightEXE']]

# Step 2: Group the dataframe by 'Origin' and 'Destination'
groups = df_od.groupby(['Origin', 'Destination'])

# Step 3: Sort the groups by 'FlightDate' in ascending order
for name, group in groups:
    group.sort_values(by='FlightDate', inplace=True)

# Step 4 and 5: Apply exponential smoothing and forecast for next 7 days
forecasts = pd.DataFrame(columns=['FlightDate', 'Origin', 'Destination', 'Forecast'])
for name, group in groups:
    data = np.array(group['ChargeableWeightEXE'])
    fit = SimpleExpSmoothing(data).fit()
    forecast = fit.forecast(7)
    last_date = pd.to_datetime(group['FlightDate'].iloc[-1])
    next_dates = pd.date_range(last_date + pd.Timedelta(days=1), periods=7, freq='D')
    next_dates = pd.DataFrame(next_dates, columns=['FlightDate'])
    next_dates['Origin'] = name[0]
    next_dates['Destination'] = name[1]
    next_dates['Forecast'] = forecast
    forecasts = forecasts.append(next_dates, ignore_index=True)

# Print the forecasted dataframe
print(forecasts)
```
